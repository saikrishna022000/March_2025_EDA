**one hot encoder**
- one hot means at a time only ine will **ON** and another is **OFF**
- **ON** represent with 1
- **OFF** represent with 0
- one hot encoder creates new columns
- for example case status has two unique lables
- number of new columns = number of unique lables 

- case status has two unique lables
- certified
- denied
- it will create two new columns
   - case_status_certified
   - case_status_denied
- original columns prefix  

 |case_status|case_status_Certified|case_status_Denied|
 |-|-|-|
 |Certified|1|0| 
 |Denied|0|1|

**Advantages**
- Generally ML models excepect independance between input variables
- there is no relation between the variables
- becase of one hot at a time only one will be on, other will off
- this indicates the new columns are independent each other
- this independence behaviour also called as **Orthogonality**
- independant means no relation
- no relation mean prependicular each other
- perpendicular means 90 degrees phase shift
- 90degrees mean orthogonal

**Disadvantage**
- it will create more columns
- more columns means more data
- more data more complexity w.r.t time and infra
- this is called as **cruse of Dimensionality**
