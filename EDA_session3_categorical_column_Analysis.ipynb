```python
import os 
new_folder='frequency_table_vc'
############# Join the paths #############
cwd=os.getcwd() 
new_dir=os.path.join(cwd,new_folder) 
########## Make the folder ############### 
try: 
    os.makedirs(new_dir)
except Exception as e: 
        print(e) 
for i in cat[1:]: 
    visa_df[i].value_counts()
    keys=visa_df[i].value_counts().keys() 
    values=visa_df[i].value_counts().values 
    cols=['Lables','Count'] 
    df=pd.DataFrame(zip(keys,values), columns=cols) 
    df.to_csv(f'{new_dir}\\{i}_table.csv')
```

**Barcharts**
- bar charts are representation of frequency tables
- it is under matplotlib package
- the method nmae is **bar**
   - plt.bar

```python
# step-1: read the data
# step-2: apply value counts
# step-3: vallues counts will give keys and values
# step-4: consider keys as x-axis
# step-5: consider values as y-axis
# step-6: pass those twodirectly in plt.bar
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.figure(figsize=(10,5))
plt.bar(keys,values)
plt.xlabel('class')   # xaxis name
plt.ylabel('class Frequency') # y-axis name
plt.title('continent Barchart')
plt.savefig('continent_bar_chat.jpg')
plt.savefig('continent_bar_chat.png')
plt.show() # show always at last
```

**count plot**
- we can plot the bar charts using seaborn package also
- the method name is **count plot**
- if we wnat tom piot bar chart using matplotlib
- we required two columns
    - class
    - class frequency
- but class and frequency is not avialable directly
- for that we are making frequency table first then plotting
- but using seaborn and count plot,weno need to make frequency table
- we can have only a column name and its data enough  

```python
order=['Asia','Europe','North America','South America','Africa','Ocenia']
keys=visa_df['continent'].value_counts().keys()
plt.figure(figsize=(10,5))
sns.countplot(data=visa_df,x='continent',order=keys)
plt.title('Bar chart')
plt.savefig('continent_countplot.jpg')
```
